```python
import pandas as pd

# Load historic data
hist_dom = pd.read_csv("2026_Final_Historic_Domestic.csv")
hist_int = pd.read_csv("2026_Final_Historic_International.csv")
sev_dom = pd.read_csv("2026_Final_Supervisory_Severely_Adverse_Domestic.csv")
sev_int = pd.read_csv("2026_Final_Supervisory_Severely_Adverse_International.csv")

#delete scenario name column
def drop_scenario_name(df):
    return df.loc[:, ~df.columns.str.contains("Scenario Name", case=False)]
    
hist_dom = drop_scenario_name(hist_dom)
hist_int = drop_scenario_name(hist_int)
sev_dom  = drop_scenario_name(sev_dom)
sev_int  = drop_scenario_name(sev_int)

# Extract date time
hist_dom["YEAR"] = hist_dom["Date"].str[:4].astype(int)
hist_int["YEAR"] = hist_int["Date"].str[:4].astype(int)
sev_dom["YEAR"] = sev_dom["Date"].str[:4].astype(int)
sev_int["YEAR"] = sev_int["Date"].str[:4].astype(int)

# Filter years 2000â€“2025
hist_dom = hist_dom[(hist_dom["YEAR"] >= 2000) & (hist_dom["YEAR"] <= 2025)]
hist_int = hist_int[(hist_int["YEAR"] >= 2000) & (hist_int["YEAR"] <= 2025)]
sev_dom = sev_dom[(sev_dom["YEAR"] >= 2026) & (sev_dom["YEAR"] <= 2029)]
sev_int = sev_int[(sev_int["YEAR"] >= 2026) & (sev_int["YEAR"] <= 2029)]

# Merge historic domestic + international
historic_merged = pd.merge(
    hist_dom.drop(columns=["YEAR"]),
    hist_int.drop(columns=["YEAR"]),
    on="Date",
    how="inner",
    suffixes=("_dom", "_int")
)

# Merge severely adverse domestic + international
severely_adverse_merged = pd.merge(
    sev_dom.drop(columns=["YEAR"]),
    sev_int.drop(columns=["YEAR"]),
    on="Date",
    how="inner",
    suffixes=("_dom", "_int")
)

# Merge historic and severe adverse forecast
severe_scenario_data = pd.concat(
    [historic_merged, severely_adverse_merged],
    axis=0,
    ignore_index=True
)

#delete scenario name column
def drop_scenario_name(df):
    return df.loc[:, ~df.columns.str.contains("Scenario Name", case=False)]
    
hist_dom = drop_scenario_name(hist_dom)
hist_int = drop_scenario_name(hist_int)
sev_dom  = drop_scenario_name(sev_dom)
sev_int  = drop_scenario_name(sev_int)

print("Final macro panel:", severe_scenario_data.shape)
severe_scenario_data

severe_scenario_data.to_csv("Severely Adverse Scenario Data.csv", index=False)
```
